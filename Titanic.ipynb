```python
titanic=pd.read_csv("train.csv")
titanic.head()
```

```python
df_sex = pd.get_dummies(titanic['Sex'],prefix='Sex')
df_sex.head()

df_class=pd.get_dummies(titanic['Pclass'],prefix='Class')


titanic = pd.concat([titanic,df_class, df_sex], axis=1)
titanic.head()

titanic=titanic.dropna(axis=0)

titanic.head()
```

```python
#This is for schoilarly purposes
#How to find nulls and nans in your pandas dataframe

#titanic.isnull()
pd.isnull(titanic)

np.where(pd.isnull(titanic))
```

```python
xLabels=[c for c in titanic.columns if c in ['Class_1','Class_2','Class_3','Sex_female','Sex_male','Age','SibSp','Parch']]
X=titanic[xLabels]
Y=titanic['Survived']
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(X,Y)
print(clf.score(X,Y))
```
